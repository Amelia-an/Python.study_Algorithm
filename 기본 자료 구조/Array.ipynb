# 파이썬과 배열

파이썬에는 리스트, 튜플과 같은 자료형을 통해 배열을 사용할 수 있음

파이썬에서는 배열의 최대길이를 지정하지 않기 때문에 데이터 추가/삭제시 편리하다

```python
country = "two hand"
print(country)

country = country + " place"
print(country)

# 문자열의 각 문자도 배열의 각 요소로 볼 수있다
# 배열의 크기를 지정하지 않는데 위 예시에서 보면"two hand" 라는 문자열에 "place" 를 쉽게 이어붙일 수 있다
```

```python
# dateset리스트에서 전체 이름 안에 'a'가 몇 번 나왔는지 출력하기
# CASE1). 문자열의 count 함수 사용(파이썬의 배열[리스트]를 사용하는 장점 중 하나인 다양한 내장함수, 그 중 하나인 'count')

dataset = ["abcabc","bbab","cccc","acscd","acacac"]

total = 0 

for data in dataset:
    count = data.count('a')
    total = total + count

    
print("a의 개수는", total)
```

```python
# CASE2). list 함수를 사용하지 못한다는 전제조건 하에

dataset = ["abcabc","bbab","cccc","acscd","acacac"]

total = 0 

for data in dataset:
    for index in range(len(data)):
        if data[index] == "a":
            total += 1

    
print("a의 개수는", total)
```

case1과 리스트를 통해 loop를 도는 부분은 같다. 다른 점은 각 문자열 요소의 길이를 구해서 문자열도 하나의 배열로 보았다는 점이다. 아래 그림처럼 각 5개의 row(행)을 가진 2차원 배열이라고 생각하면 될 것 같다.
